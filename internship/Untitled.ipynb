```python
import pandas as pd
import numpy as np
fund_file = pd.read_csv('Nippon_India_Large_Cap_Fund.csv')
# fund_file.drop(['Repurchase Price', 'Sale Price'], axis = 1, inplace = True)

benchmark = pd.read_csv("Benchmark_BSE_100.csv")
# benchmark.drop(columns = ['Open', 'Low', 'High'], inplace = True)

index_date = np.array([])
fund_close = np.array([])
bench_close = np.array([])
for i in range(len(fund_file)):
#     ind = fund_file.index[i]
        index_match = None
        try:
            index_match = benchmark.loc[benchmark.Date == fund_file['NAV date'].iloc[i]].index
            bench_close = np.append(bench_close, benchmark['Close'].iloc[index_match[0]])
            index_date = np.append(index_date, fund_file['NAV date'].iloc[i])
            fund_close = np.append(fund_close, fund_file['Net Asset Value'].iloc[i])
        except:
            pass

percent_fund = np.array([0])
percent_bench = np.array([0])
for i in range(1, len(fund_close)):
    
    percent_fund = np.append(percent_fund, (fund_close[i]-fund_close[0])/fund_close[0]*100)

    percent_bench = np.append(percent_bench, (bench_close[i]-bench_close[0])/bench_close[0]*100)

data1 = {'percent_fund': percent_fund, 'percent_bench': percent_bench}
df1 = pd.DataFrame(data1, index = index_date)
df1
```
